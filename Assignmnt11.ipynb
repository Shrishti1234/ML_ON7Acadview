Create a new column 'Has_Cabin' that tells you whether passengers had a cabin or not

```python
# creating new column named as Has_cabin with null values
df['Has_cabin']=np.nan
df.head()
```

```python
# Filling values in column Has_cabin on the basis whether passnegers had cabin or not
i=0
for data in df['Cabin']:
    if pd.isnull(data):
        df.iloc[i,12]=0
        
    else:
        df.iloc[i,12]=1
    i=i+1
    
```

#### Ques:4 

Creating new column named as Title ,which contain the titles of passengers as Mr.,Mrs.,Miss,Master.and Special

```python
# Creating column title with null values
df['Title']=np.nan
df.head()
```

```python
# Use of regular expression to fill the values in column Title
i=0
import re
for data in df['Name']:
    if(re.search(r'Mr.',data)):
        if(re.search(r'Mrs.',data)):
            df.iloc[i,13]="Mrs."
        else:
            df.iloc[i,13]="Mr."
    elif(re.search(r'Miss.',data)):
            df.iloc[i,13]='Miss.'
        
    elif(re.search(r'Master.',data)):
            df.iloc[i,13]="Master."
    else:
        df.iloc[i,13]="Special"
    
    i=i+1   

        
```

#### Ques:5

Input the missing values using fillna method

```python
# Filling null values in age column with mean value
df=df.fillna(df.mean())
# Filling null values in column Cabin and Embarked with "0"
df.Cabin.fillna('0',inplace=True)
df.Embarked.fillna('0',inplace=True)
```

 Create two new columns “Categorical_age” and “Categorical_fare”, where you will put people that are within a certain range of age or fare in the same bin.

#### Ques:8
Transforming all variables into numeric ones

```python
# Converting string object into numerical values using mapping
map1 = {k: v for v, k in enumerate(df.Sex.unique())}
print(map1)
map2 = {k: v for v, k in enumerate(df.Embarked.unique())}
print(map2)
```

```python
# Creating New columns which have numerical values for Sex and embarked variables
df['Sex_Numeric'] = df.Sex.map(map1)
df['Embarked_numeric']=df.Embarked.map(map2)
```
